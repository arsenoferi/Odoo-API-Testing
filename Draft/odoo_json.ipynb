```python
import requests
import json
import pandas as pd

# Odoo server information
```

```python
url = 'https://bunker2.odoo.com/jsonrpc'
db = 'bunker2'
username = 'bizops@bunkertech.id'
password = 'Bunker354'

```

```python
headers = {'Content-Type': 'application/json'}

# Authentication payload
payload = {
    "jsonrpc": "2.0",
    "method": "call",
    "params": {
        "service": "common",
        "method": "login",
        "args": [db, username, password],
    },
    "id": 1,
}

response = requests.post(url, data=json.dumps(payload), headers=headers)
uid = response.json().get("result")

if uid:
    print("Successfully authenticated. UID:", uid)
else:
    print("Authentication failed.")
```

```python
data_payload = {
    "jsonrpc": "2.0",
    "method": "call",
    "params": {
        "service": "object",
        "method": "execute_kw",
        "args": [
            db, uid, password,
            'res.partner',  # Model
            'search_read',  # Method
            [],  # Domain (empty array for all records)
            {'fields': ['name', 'email', 'phone']}  # Fields to fetch
        ],
    },
    "id": 2,
}
response = requests.post(url, data=json.dumps(data_payload), headers=headers)
data = response.json().get("result", [])

print("Fetched data:", data)
```
